```python
#Import Packages
import numpy as np
import pandas as pd
import datetime
import random
from matplotlib import pyplot as plt
import copy
from collections import Counter
%matplotlib inline
SEED = 213
random.seed(SEED)
```

```python
# Quicksort implementation from textbook
def quick_sort1(array):
  if len(array) < 2:
    # base case, arrays with 0 or 1 element are already "sorted"
    return array
  else:
    # recursive case
    pivot = array[0]
    # sub-array of all the elements less than the pivot
    less = [i for i in array[1:] if i <= pivot]
    # sub-array of all the elements greater than the pivot
    greater = [i for i in array[1:] if i > pivot]
    return quick_sort1(less) + [pivot] + quick_sort1(greater)
```

```python
# In-place QuickSort implementation 
# https://www.geeksforgeeks.org/quick-sort/
  
# This Function handles sorting part of quick sort
# start and end points to first and last element of
# an array respectively
def partition(start, end, array):
      
    # Initializing pivot's index to start
    pivot_index = start 
    pivot = array[pivot_index]
      
    # This loop runs till start pointer crosses 
    # end pointer, and when it does we swap the
    # pivot with element on end pointer
    while start < end:
          
        # Increment the start pointer till it finds an 
        # element greater than  pivot 
        while start < len(array) and array[start] <= pivot:
            start += 1
              
        # Decrement the end pointer till it finds an 
        # element less than pivot
        while array[end] > pivot:
            end -= 1
          
        # If start and end have not crossed each other, 
        # swap the numbers on start and end
        if(start < end):
            array[start], array[end] = array[end], array[start]
      
    # Swap pivot element with element on end pointer.
    # This puts pivot on its correct sorted place.
    array[end], array[pivot_index] = array[pivot_index], array[end]
     
    # Returning end pointer to divide the array into 2
    return end
      
# The main function that implements QuickSort 
def quick_sort2(start, end, array):
      
    if (start < end):
          
        # p is partitioning index, array[p] 
        # is at right place
        p = partition(start, end, array)
          
        # Sort elements before partition 
        # and after partition
        quick_sort2(start, p - 1, array)
        quick_sort2(p + 1, end, array)
```

```python
#Bubble Sort (provided in starter script)
def bubble_sort(xlist):
    n = len(xlist)
    
    #Traverse array
    for i in range(n):
        # Last i elements are already in place
        for j in range(0, n-i-1):
            #Traverse array from 0 to n-i-1
            #Swap if element found is greater than next element
            if xlist[j] > xlist[j+1]:
                xlist[j], xlist[j+1] = xlist[j+1], xlist[j]
    return xlist
```

```python
# Insertion sort
# https://www.geeksforgeeks.org/insertion-sort/
def insertion_sort(arr):
 
    # Traverse through 1 to len(arr)
    for i in range(1, len(arr)):
 
        key = arr[i]
 
        # Move elements of arr[0..i-1], that are
        # greater than key, to one position ahead
        # of their current position
        j = i - 1
        while j >= 0 and key < arr[j]:
                arr[j + 1] = arr[j]
                j -= 1
        arr[j + 1] = key
```

```python
# verify for the selected algorithms whether sorting is in-place or not

# quicksort1 (from textbook)
arr = [8, 11, 7, -3, 5, -9]
print('quicksort1 (not in-place)')
print('arr before:', arr)
print('function output:', quick_sort1(arr)) # sorted
print('arr after:', arr) # not sorted in-place

# quicksort2 (sorted in-place)
arr = [8, 11, 7, -3, 5, -9]
print('\nquicksort2 (in-place)')
print('arr before:', arr)
print('function output:', quick_sort2(start=0, end=len(arr)-1, array=arr)) # function does not return arr
print('arr after:', arr) # sorted in-place

# bubble sort
arr = [8, 11, 7, -3, 5, -9]
print('\nbubble sort')
print('arr before:', arr)
print('function output:', bubble_sort(arr)) # sorted
print('arr after:', arr) # sorted in-place

# insertion sort
arr = [8, 11, 7, -3, 5, -9]
print('\ninsertion sort')
print('arr before:', arr)
print('function output:', insertion_sort(arr)) # function does not return arr
print('arr after:', arr) # sorted in-place
```

#### Quicksort
* Quicksort uses the divide & conquer strategy, which involves defining a base case, then reducing the problem recursively to get to the base case. 


* Here, the base case is an array with length < 2 (nothing to sort). While not the base case, partitioning is performed with the following steps: (1) pick a pivot; (2) partition the array into two subarrays: elements less than the pivot and elements greater than the pivot; and (3) call quicksort recursively on the two subarrays.


* In the worst case, quicksort takes O(n<sup>2</sup>) time (height of call stack = O(n); elements per call stack level = O(n)). In the best case, quicksort takes O(n log n) time (height of call stack = O(log n); elements per call stack level = O(n)). However, "the best case is also the average case. If you always choose a random element in the array as the pivot, quicksort will complete in O(n log n) time on average" (ref. 1).


* Two implementations of quicksort were tested here to compare the effect of in-place sorting on execution time:
    * Quicksort1 (ref. 1), which utilizes additional arrays to perform sorting (e.g., `left` and `right`) and thus does not sort in-place (needing to borrow space/memory).
    
    * Quicksort2 (ref. 2), which could be considered to sort in-place, as it uses extra space only for storing recursive function calls but not for manipulating the input.
    
#### Comparing Quicksort with Bubble sort and Insertion sort
* Bubble sort works by repeatedly swapping the adjacent elements if they are in the wrong order. The worst and average time complexity is O(n<sup>2</sup>). Bubble sort sorts in-place.


* Insertion sort works by splitting the array into sorted and unsorted subarrays. At each iteration, an element from the unsorted subarray is picked and placed/inserted at the correct position in the sorted subarray, with larger values in the sorted subarray shifting up to make space. The worst and average time complexity is O(n<sup>2</sup>). Insertion sort sorts in-place. 


* Note that randomly generated, unsorted lists (as described in Requirement 1) were used as inputs in this experiment. Algorithm performance on already-sorted lists are not considered here.
    * The time taken by quicksort depends on the input and partition strategy. Quicksort will run at the worst case of O(n<sup>2</sup>) if the smallest or largest element is always picked as the pivot. This would happen, for example, if it always picks the last element as the pivot and the list is already sorted.
    * The time taken by bubble sort and insertion sort also depends on the input array. For example, bubble sort and insertion sort run at the best case O(n) when the list is already sorted.

#### Results
* Time complexity (see Table and Figure)
    * In terms of performance, Quicksort1 (not sorted in-place) > Quicksort2 (sorted in-place) > Insertion sort > Bubble sort across the list lengths tested. The difference in execution time increased with increasing list length. 
    
    * Both implementations of quicksort greatly outperformed insertion sort and bubble sort. Insertion sort outperformed bubble sort even though both are time complexity O(n<sup>2</sup>).


* Space complexity
    * Insertion sort and bubble sort both sort in-place and do not need additional memory (beyond that used by the list itself) for sorting or stack space for recursive calls. 
    
    * Quicksort2, which sorts in-place, requires O(log n) stack space in the average case (or O(n) in the worst case). 
    
    * Quicksort1, which had the best execution time, does not sort in-place and requires additional memory for the `left` and `right` arrays, along with O(log n) stack space in the average case. 
    * Thus, there is a tradeoff between the algorithms in terms of balancing execution time with space/memory usage (further discussed below).


#### Conclusions 
* This experiment confirmed that quicksort is one of the fastest sorting algorithms out there (ref. 1). Quicksort hits the average case way more often than the worst case (ref. 1). With an average time complexity of O(n log n) and a space complexity of O(log n) stack space (with in-place sorting), quicksort would be a useful tool for data engineers. However, data engineers should also be aware of the scenarios where quicksort degrades to the worst-case time complexity of O(n<sup>2</sup>), and optimize the partition strategy accordingly.


* Regarding Quicksort1, which does not sort in-place, we will likely see no (or negligible) negative impact on space/memory usage for small datasets on the order of megabytes (e.g., 1-100 MB). However, for datasets on the order of gigabytes (e.g., 1 GB+), borrowing the extra space/memory for sorting will become more challenging, depending on available computing resources. The choice between Quicksort1 and Quicksort2 may therefore depend on the size of the dataset.


* Insertion sort ran faster than bubble sort and does not appear to be any more complex than bubble sort. Thus, insertion sort could be an option if we wish to use as little memory as possible. As a non-recursive sorting algorithm, insertion sort could even be faster than quicksort for very small arrays (not tested in this experiment; needs to be verified). If this is true, one hybrid strategy could be to use a non-recursive sorting algorithm such as insertion sort for very small arrays and switching to quicksort as array length increases. This strategy may allow data engineers to take advantage of the algorithms' relative strengths.
